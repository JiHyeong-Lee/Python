```python
# 데이터 프레임 설정 시 .copy()를 절대 빼먹지 말 것
df_seoul_hospital = df[(df["상권업종소분류명"] == "종합병원") & 
                       (df["시도명"] == "서울특별시")].copy()
df_seoul_hospital
```

```python
df_seoul_hospital["시군구명"].value_counts()
```

```python
# 텍스트 데이터 색인하기
# 색인 전 상호명 중에 종합병원이 아닌 데이터를 찾기
# .str.contains("") 텍스트에 포함되어 있는 지 찾아주는 함수
# 앞에 ~표시를 하면 안 들어간 것만 색인됨. 즉, 반대로 됨

df_seoul_hospital[~df_seoul_hospital["상호명"].str.contains("종합병원")]
```

```python
# 상호명에 종합병원이 들어가지 않는 상호명들을 색인하고, 그 상호명들이 무엇인지 알아보자
df_seoul_hospital.loc[~df_seoul_hospital["상호명"].
                      str.contains("종합병원"), "상호명"].unique()
```

```python
# 꽃배달이 상호명에 들어가는 것들을 찾아보자
df_seoul_hospital[df_seoul_hospital["상호명"].str.contains("꽃배달")]
```

```python
# 상호명에 의료기가 들어가는 데이터를 찾아보자
df_seoul_hospital[df_seoul_hospital["상호명"].str.contains("의료기")]
```

```python
# 꽃배달/의료기/장례식장/상담소/어린이집은 종합병원과 무관하기 때문에
# 전처리를 위해서 해당 텍스트를 한 번에 검색하기
# 텍스트를 or로 넣고 한 꺼번에 ""에 넣어줘야 한다.

df_seoul_hospital[df_seoul_hospital["상호명"]
                  .str.contains("꽃배달|의료기|장례식장|상담소|어린이집")]
```

```python
# 인덱스 값만 추출
df_seoul_hospital[df_seoul_hospital["상호명"]
                  .str.contains("꽃배달|의료기|장례식장|상담소|어린이집")].index
```

```python
# 인덱스 값을 drop_row라는 변수에 할당
drop_row = df_seoul_hospital[df_seoul_hospital["상호명"]
                  .str.contains("꽃배달|의료기|장례식장|상담소|어린이집")].index

# 인덱스 값을 리스트로 변환
drop_row = drop_row.tolist()
drop_row
```

```python
# 의원으로 끝나는 데이터도 인덱스를 찾아서 drop_row2 에 담아주고 list로 만들자
# 의원으로 끝나는 것은 str.endswith("")함수를 사용

drop_row2 = df_seoul_hospital[df_seoul_hospital["상호명"]
                  .str.endswith("의원")].index

drop_row2 = drop_row2.tolist()
drop_row2
```

```python
# 삭제할 행을 drop_row 에 합치기

drop_row = drop_row + drop_row2
len(drop_row)
```

```python
# 해당 셀을 삭제하고 삭제 전과 후의 행의 갯수를 비교하자
# .drop(행의 index)를 사용해서 삭제할 수 있음

print(df_seoul_hospital.shape)
df_seoul_hospital = df_seoul_hospital.drop(drop_row, axis=0)
print(df_seoul_hospital.shape)
```

```python
# plt.figure(figsize=(15,4))를 통해서 사이즈를 조정
# seaborn을 사용하는 경우!!!

plt.figure(figsize=(15,4))
sns.countplot(data=df_seoul_hospital, x="시군구명")
```

```python
# order를 통해서 차트에 나타나는 순서를 지정해 줄 수 있음

plt.figure(figsize=(15,4))
sns.countplot(data=df_seoul_hospital, x="시군구명"
             , order = df_seoul_hospital["시군구명"].value_counts().index)
```

```python
# 지도의 중심을 지정하기 위해 위도와 경도의 평균을 구합니다.

df_seoul_hospital["위도"].mean()
df_seoul_hospital["경도"].mean()
```

```python
map = folium.Map(location = [df_seoul_hospital["위도"].mean(), 
                       df_seoul_hospital["경도"].mean()], zoom_start = 12)

for n in df_seoul_hospital.index :
    name = df_seoul_hospital.loc[n, "상호명"]
    adress = df_seoul_hospital.loc[n, "도로명주소"]
    #문자열 formating 기법 중 하나
    popup = f"{name}-{adress}"
    location = [df_seoul_hospital.loc[n,"위도"], df_seoul_hospital.loc[n, "경도"]]
    folium.Marker(location = location,
                 popup=popup,
                 ).add_to(map)
    
map
    
```
